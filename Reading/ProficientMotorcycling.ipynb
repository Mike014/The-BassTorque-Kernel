# CAPITOLO 2: MOTORCYCLE DYNAMICS - WHAT KEEPS IT BALANCED?

## IL PROBLEMA: TRE RIDER INCOMPETENTI

**Drifting Dan:**
- Tourer pesante, vuole curva crisp da fermo
- Moto "prende comando" e varca centerline
- Panic → tira i manubri (SBAGLIATO) → varca ancora di più

**Wandering Wanda:**
- Cruiser, vuole stare in mezzo corsia
- Moto "scivola" verso bordo, poi centerline
- Ha paura del contrasterzo, non lo tenta

**Beemer Bob:**
- Sport tourer, va bene in velocità
- Parcheggio = lotta costante contro i fender delle auto
- Non sa che il GRIP, non le ginocchia, controlla la direzione

### Il Problema Comune

> "Non capiscono COME le moto sterzano e si equilibrano."

Dan pensa di sterzare i manubri (sbagliato).
Wanda ha paura di provare.
Bob spinge le ginocchia (inefficace).

---

## LA SOLUZIONE SEMPLICE: COUNTERSTEERING

**La Regola:**
```
Per piegarti a DESTRA:
→ SPINGI il grip DESTRO

Per piegarti a SINISTRA:
→ SPINGI il grip SINISTRO
```

**Perché si chiama "Counter":**
Momentaneamente sterzate il FRONT WHEEL OPPOSTO (counter) 
a dove volete piegarvi.

### Esempio Pratico

**Scenario:** Moto si avvicina a auto parcheggiata a DESTRA.
```
❌ SBAGLIATO (cosa fa Dan):
Tira grip destro (spavento)
→ Front wheel gira a destra
→ Moto scende ancora più a destra
→ Crash!

✅ CORRETTO (contrasterzo):
Spingi grip SINISTRO
→ Front wheel momentaneamente gira a sinistra
→ Moto cade a SINISTRA (lontano da auto)
→ Eviti!
```

---

## I DUE SEGRETI PER CONTROLLARE

### Segreto #1: COUNTERSTEERING
- Spingi il grip nella direzione dove vuoi piegarti
- Non tirare
- Non muscolare

### Segreto #2: GUARDA DOVE VUOI ANDARE
- Non guardare il pothole → lo colpisci
- Guarda un lato della strada → ci vai
- Non fissare la centerline → la varchi

**Nota:** Anche novice rider che non sanno il countersteering
guadagnano controllo solo guardando bene.

---

## SITUAZIONI HAZARDOUS: IL CROSSWIND

**Scenario:** Bob esce da tunnel, colpisce crosswind 50 knot da DESTRA.

**Cosa Succede:**
- Vento spinge moto verso centerline
- È una **spinta passiva**, non una curva

**Cosa Deve Fare Bob:**
```
1. Spingi HARD sul grip DESTRO (upwind)
2. Moto si piega upwind
3. Rimani in linea retta (non curvare)
4. Mantieni pressione sul grip
```

**Perché è difficile:**
- La moto si piega MA non gira
- Steering non si "sente normale"
- Bob deve **premere sui grips**, non pensare "lean"
