La geometria dello sterzo è fondamentale per definire il comportamento dinamico di una motocicletta. I due parametri principali che determinano il bilanciamento tra agilità e stabilità sono il Rake e il Trail.

![Geometria dello sterzo: Rake e Trail](https://cdn-fastly.motorcycle.com/media/2023/02/23/8864239/what-s-the-difference-between-rake-and-trail.jpg?size=720x845&nocrop=1)

### 1. Rake (Inclinazione del Canotto di Sterzo)

Il **Rake** è l'angolo formato dall'asse dello sterzo (la linea che passa attraverso il canotto di sterzo) rispetto a una linea verticale immaginaria che scende perpendicolare al terreno.

* **Valori Tipici:** Le moto sportive hanno solitamente angoli di rake ridotti (circa 23°-25°) per ottenere una risposta fulminea. Le Cruiser o le Scrambler tendono ad avere angoli più aperti (27°-32°).
* **Effetto sulla Guida:** Un angolo di rake più ampio aumenta l'interasse della moto, rendendola estremamente stabile alle alte velocità o sui rettilinei, ma richiede più sforzo fisico per l'inserimento in curva.

### 2. Trail (Avancorsa)

Il **Trail** è la distanza orizzontale misurata a terra tra due punti:

1. Il punto di contatto dello pneumatico anteriore con il terreno.
2. Il punto in cui l'estensione dell'asse dello sterzo interseca il terreno.

* **Funzione Auto-allineante:** Il trail agisce in modo simile alle ruote di un carrello della spesa. Quando la ruota si sposta lateralmente, la distanza del trail crea una forza che tende a riportare la ruota al centro.
* **Effetto sulla Guida:** Più il trail è lungo, più la moto sarà stabile e tenderà a mantenere la traiettoria rettilinea. Un trail corto rende lo sterzo molto leggero e reattivo, ideale per i cambi di direzione rapidi, ma aumenta il rischio di instabilità o vibrazioni (shimmy) alle alte velocità.

### Relazione tra Rake e Trail

Sebbene siano due misure diverse, sono strettamente collegate. In generale:

* Aumentando l'angolo di Rake, aumenta solitamente anche il Trail.
* Le piastre di sterzo (offset) vengono utilizzate dai progettisti per regolare il Trail indipendentemente dal Rake, permettendo di affinare il feedback dello sterzo senza dover modificare strutturalmente il telaio.
