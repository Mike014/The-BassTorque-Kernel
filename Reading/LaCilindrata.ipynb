# **Cilindrata, Geometria e Forza**

## 1. Il Cuore del Cilindro: Alesaggio e Corsa

## Cos'è la Cilindrata

**Definizione:**
La cilindrata è il **volume totale d'aria spostato da TUTTI i pistoni** durante un ciclo completo (da PMI a PMS).

**Unità di Misura:**
- Centimetri Cubi (cc)
- Millilitri (mL)
- Litri (L)
           
### Definizioni Base

**Alesaggio (Bore):**
- Diametro interno del cilindro
- Più grande = valvole più grandi = più aria
- Influisce sulla velocità del motore

**Corsa (Stroke):**
- Distanza che percorre il pistone (PMI → PMS)
- Più lunga = più forza ai bassi giri
- Influisce sulla coppia

### Analogia della Siringa 
```
Siringa da cucina:
- Larghezza siringa = Alesaggio
- Quanto tiri lo stantuffo = Corsa
- Liquido totale = Cilindrata

Motore Superquadro: Largo, corsa breve → Veloci, giri alti
Motore Sottoquadro: Stretto, corsa lunga → Forti, coppia bassa
```

## 2. La Matematica della Cilindrata

### Formula
```
Cilindrata = π/4 × D² × C × N

Dove:
- D = Alesaggio (diametro)
- C = Corsa
- N = Numero cilindri
```

### Esempio Reale: Royal Enfield 650 Twin
```
D = 84 mm
C = 90 mm
N = 2 cilindri

Risultato: 648 cc
(Marketing arrotonda a 650)
```

## 3. I "Centimetri Perduti": Perché 600 ≠ 600

### Tre Ragioni Principali

**1. Soglie Fiscali:**
- Molti paesi tassano più sopra i 600cc
- I produttori restano a 580-590cc per risparmiare tasse
- Nome commerciale = 600, cilindrata reale = 580

**2. Limiti Sportivi:**
- Le gare hanno limiti rigidi (es. max 599cc)
- I costruttori non rischiano il limite massimo

**3. Efficienza Volumetrica:**
- Anche con 650cc di spazio, il motore ne riempie solo il 75-85%
- Cause: filtri dell'aria, resistenza valvole, frizione miscela

## 4. Coppia vs Potenza: I "cc" Non Sono Tutto

### Definizioni Critiche

**Coppia (Torque):**
- Forza pura applicata all'albero motore
- Quella spinta che ti attacca al sedile ai bassi giri
- **Analogia:** Usare una chiave inglese per svitare un bullone con forza

**Potenza (Power):**
- Velocità con cui applichi la forza
- Vincola la velocità massima
- **Analogia:** Svitare il bullone velocemente

### Confronto Pratico
```
ALTA COPPIA (es. Cruiser 650cc):
- Gigante che porta 10 pacchi pesanti
- Cammina lentamente
- Forte accelerazione ai bassi giri
- Bassa velocità massima

ALTA POTENZA (es. Sportiva 600cc):
- Atleta che porta 1 pacco alla volta
- Corre velocissimamente
- Accelerazione media
- Alta velocità massima
```
## **Conclusione**

Una 650cc Cruiser può avere più "forza" (coppia) di una 600cc sportiva, ma la sportiva avrà velocità massima superiore grazie alla potenza.

## **Il Calcolo (Passo Dopo Passo):**

### **Royal Enfield 650 Twin (2 cilindri):**

```
1. Alesaggio D = 84 mm (diametro cilindro)
2. Corsa C = 90 mm (movimento pistone)
3. Numero cilindri N = 2

Cilindrata = π/4 × 84² × 90 × 2
           = 0.785 × 7,056 × 90 × 2
           = 648 cc
```

**Cosa significa 648 cc?**

Ogni volta che i 2 pistoni completano un ciclo (su + giù), **spostano 648 millilitri d'aria totale**.

## **L'Analogia Corretta (Siringa):**

```
Prendi una siringa medica:

Alesaggio = Larghezza della siringa (es. 2 cm diametro)
Corsa = Quanto tiri lo stantuffo (es. 10 cm)

Cilindrata = Volume aspirato
           = Area × Corsa
           = (π/4 × 2²) × 10
           = 31.4 cm³

Se hai 2 siringhe uguali: Cilindrata totale = 62.8 cm³
```

## **La Differenza (Riassunto):**

| Termine | Definizione | Esempio |
|---------|------------|---------|
| **Alesaggio (D)** | Diametro del cilindro | 84 mm |
| **Corsa (C)** | Distanza movimento pistone | 90 mm |
| **Cilindrata** | Volume totale spostato | 648 cc = 648 mL |
| **Giri** | Numero di rotazioni albero motore | 6,000 RPM |
